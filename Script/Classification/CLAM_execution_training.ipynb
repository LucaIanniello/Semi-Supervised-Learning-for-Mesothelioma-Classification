```python
if feature_extractor in ["resnet50-trident", "phikon-trident", "univ1-trident", "univ2-trident"]:
    import h5py
    import numpy as np
    folder_dest = "./results_features/pt_files/"
    B = []
    E = []
    S = []

    if feature_extractor == "resnet50-trident":
        folder_h5_B = "./results_features/pt_files/trident_processed_resnet/B/20x_256px_0px_overlap/features_resnet50/"
        folder_h5_E = "./results_features/pt_files/trident_processed_resnet/B/20x_256px_0px_overlap/features_resnet50/"
        folder_h5_S = "./results_features/pt_files/trident_processed_resnet/B/20x_256px_0px_overlap/features_resnet50/"
    elif feature_extractor == "phikon-trident":
        folder_h5_B = "./results_features/pt_files/trident_processed_phikon/B/20x_224px_0px_overlap/features_phikon_v2/"
        folder_h5_E = "./results_features/pt_files/trident_processed_phikon/E/20x_224px_0px_overlap/features_phikon_v2/"
        folder_h5_S = "./results_features/pt_files/trident_processed_phikon/S/20x_224px_0px_overlap/features_phikon_v2/"
        
    elif feature_extractor == "univ1-trident":
        folder_h5_B = "./results_features/pt_files/trident_processed_univ1/B/20x_256px_0px_overlap/features_uni_v1/"
        folder_h5_E ="./results_features/pt_files/trident_processed_univ1/E/20x_256px_0px_overlap/features_uni_v1/"
        folder_h5_S = "./results_features/pt_files/trident_processed_univ1/S/20x_256px_0px_overlap/features_uni_v1/"
    elif feature_extractor == "univ2-trident":
        folder_h5_B = "./results_features/pt_files/trident_processed_univ2/B/20x_256px_0px_overlap/features_uni_v2/"
        folder_h5_E ="./results_features/pt_files/trident_processed_univ2/E/20x_256px_0px_overlap/features_uni_v2/"
        folder_h5_S = "./results_features/pt_files/trident_processed_univ2/S/20x_256px_0px_overlap/features_uni_v2/"
        
    for fname in os.listdir(folder_h5_B):
            if not fname.endswith('.h5'):
                continue
            file_h5 = os.path.join(folder_h5_B, fname)
            with h5py.File(file_h5, 'r') as f:
                features_np = f['features'][:]
            fname = fname.split('.')[0]
            features = torch.from_numpy(features_np).float()
            torch.save(features, folder_dest + fname + ".pt")
            B.append(fname)
    for fname in os.listdir(folder_h5_E):
            if not fname.endswith('.h5'):
                continue
            file_h5 = os.path.join(folder_h5_E, fname)
            with h5py.File(file_h5, 'r') as f:
                features_np = f['features'][:]
            fname = fname.split('.')[0]
            features = torch.from_numpy(features_np).float()
            torch.save(features, folder_dest + fname + ".pt")
            E.append(fname)
    for fname in os.listdir(folder_h5_S):
            if not fname.endswith('.h5'):
                continue
            file_h5 = os.path.join(folder_h5_S, fname)
            with h5py.File(file_h5, 'r') as f:
                features_np = f['features'][:]
            fname = fname.split('.')[0]
            features = torch.from_numpy(features_np).float()
            torch.save(features, folder_dest + fname + ".pt")
            S.append(fname)
            
    dataset_csv = []
    for i in range(len(B)):
        v = B[i].split('.')[0]
        dataset_csv.append([v,v,'B'])
    for i in range(len(S)):
        v = S[i].split('.')[0]
        dataset_csv.append([v,v,'S'])
    for i in range(len(E)):
        v = E[i].split('.')[0]
        dataset_csv.append([v,v,'E'])

    import csv
    # Nome del file CSV
    file_csv = folder_dest + 'datasetComposition.csv'

    # Scrittura del file CSV
    with open(file_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        # Scrivere l'intestazione
        writer.writerow(['case_id', 'slide_id', 'label'])
        # Scrivere i dati
        writer.writerows(dataset_csv)

    print(f"File CSV salvato: {file_csv}")
```
